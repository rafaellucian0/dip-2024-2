```python
import matplotlib.pyplot as plt
import cv2
import numpy as np
from skimage.color import rgb2gray, label2rgb
from skimage.segmentation import chan_vese, slic
```

```python
def chan_image(img):
  fig, axes = plt.subplots(1, 3, figsize=(10, 10))

  image = cv2.imread(img)
  gray = rgb2gray(image)

  chanvese_gray = chan_vese(gray, max_num_iter=100, extended_output=True)

  ax = axes.flatten()

  ax[0].imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
  ax[0].set_title("Original Image")

  ax[1].imshow(chanvese_gray[0], cmap="gray")
  ax[1].set_title("Chan-Vese segmented - 100 iterations")

  ax[2].imshow(chanvese_gray[1], cmap="gray")
  ax[2].set_title("Final Level Set")
  plt.show()
```

```python
def slic_image(img):
  fig, axes = plt.subplots(1, 2, figsize=(10, 10))

  image = cv2.imread(img)

  image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
  image_segments = slic(image_rgb, n_segments=50, compactness=10)

  ax = axes.flatten()

  ax[0].imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
  ax[0].set_title("Original Image")

  ax[1].imshow(label2rgb(image_segments, image_rgb, kind='avg'))
  ax[1].set_title("SLIC segmented Image")
  plt.show()
```
